```python
# Import libraries
import requests
import pandas as pd
```

```python
# Get everything
all = requests.get('http://localhost:9000/all/')
df1 = pd.DataFrame.from_dict(all.json())
df1.head()
```

```python
# Get everything from final episode
finale = requests.get('http://localhost:9000/episode/finale/')
df2 = pd.DataFrame.from_dict(finale.json())
df2.head()
```

```python
# Get everything from Jim (character) 
jim = requests.get('http://localhost:9000/character/jim/')
df3 = pd.DataFrame.from_dict(jim.json())
df3.head()
```

```python
# Get all characters and their total lines
characters = requests.get('http://localhost:9000/characters/')
df3 = pd.DataFrame.from_dict(characters.json())
df3.head()
```

```python
# Get just dialogues from episode 5, season 1
episode_5 = requests.get('http://localhost:9000/lines/episode/basketball/')
df4 = pd.DataFrame.from_dict(episode_5.json())
df4.head()
```

```python
# Get the sentiment analysis of each line from episode 5, season 1
sa_episode_5 = requests.get('http://localhost:9000/sa/episode/basketball/')
df5 = pd.DataFrame.from_dict(sa_episode_5.json())
df5.head()
```

```python
# Get just dialogues from Pam (character)
pam = requests.get('http://localhost:9000/lines/character/pam/')
df6 = pd.DataFrame.from_dict(pam.json())
df6.head()
```

```python
# Get the sentiment analysis of each line from Pam (character)
sa_pam = requests.get('http://localhost:9000/sa/character/pam/')
df7 = pd.DataFrame.from_dict(sa_pam.json())
df7.head()
```
